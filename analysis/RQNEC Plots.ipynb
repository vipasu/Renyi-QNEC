```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
sns.set_context("paper")
sns.set(font_scale=2, rc={'text.usetex' : True, 'font.family' : 'serif'}, style='white')
```

# TFIM SRDs

```python
N = 16
df = pd.read_csv("../data/WZW_{}_srd.csv".format(N))

x_range = np.arange(1, len(df.columns))/N
mask = np.where(x_range <= .5)[0]
x_range = x_range[mask]
srds = df[:-1].values
s_rel = df.iloc[-1].values[1:]
```

```python
alpha = .5
with sns.color_palette("GnBu_d", len(srds)):
    plt.figure(figsize=(8,8))
    for data in srds[::-1]:
        n, *srd = data
        srd = np.array(srd)[mask]
        plt.scatter(x_range, srd, label="$n={}$".format(n))
        plt.plot(x_range, srd, linestyle='--', alpha=alpha)
    plt.plot(x_range, s_rel[mask], marker='.', markersize=20, linestyle='--', label="$S_{rel}$")
    plt.legend(loc='best', fontsize=17)
plt.title("Critical TFIM ($L={}$)".format(N))
plt.xlabel('$x/L$')
plt.ylabel(r"$S_n(\rho||\sigma)$")
plt.savefig("../plots/TFIM_SRD.pdf", bbox_inches='tight')
```

# CFT Convergence

```python
def extract_s_rel(N):
    df = pd.read_csv("../data/tfim_{}_srd.csv".format(N))
    x_range = np.arange(1, len(df.columns))/N
    s_rel = df.iloc[-1].values[1:]
    mask = np.where(x_range <= .5)[0]
    return x_range[mask], s_rel[mask]
```

```python
sizes = [8, 12, 16, 20]
rel_ents = [extract_s_rel(N) for N in sizes]
```

```python
def analytical_s_rel(x):
    return .25 * (1 - np.pi * x /np.tan(np.pi * x))
```

```python
plt.figure(figsize=(8,8))
markers = ['v', 's', '*', 'd']
with sns.color_palette("GnBu_d", 4):
    for (i, (x, s_rel)) in enumerate(rel_ents):
        plt.scatter(x, s_rel, marker=markers[i], s=50, label="$L={}$".format(sizes[i]))

x = np.linspace(0.01,.55, 50 )        
plt.plot(x, analytical_s_rel(x), label=r'$\frac{1}{4}(1 - \pi x \cot(\pi x))$', ls='--')
plt.legend(loc='best')
plt.ylabel(r"$S_{rel}(\rho||\sigma)$")
plt.xlabel('$x/L$')
plt.title("Relative Entropy vs System Size")
plt.savefig("../plots/CFT_Convergence.pdf", bbox_inches='tight')
```
